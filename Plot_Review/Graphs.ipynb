```python
# What will be needed.
from matplotlib import pyplot as plt
```

### Simple Line Chart.

Representing the nominal GDP over time: 

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] # x-coords
gdp = [310, 520, 1100, 3000, 6000, 10290, 14900]   # y-coords

plt.plot(years, gdp, c='green', marker='o', linestyle='solid')
plt.title('Nominal GDP')
plt.ylabel('Billions of $')
plt.show()
```

### Simple Bar Chart

Representing my favorite clothing brands over (unique) quantity owned:

```python
brands=["Supreme", "Gucci", "Off-White", "Chanel", "Playboy"] # x-coords
quants=[11, 2, 6, 2, 7] # y-coords

plt.bar(range(len(brands)), quants)
plt.title('Clothing brands')
plt.ylabel('Quantity owned')
plt.xticks(range(len(brands)), brands)
plt.show()
```

Histogram for exam score distributions using bar chart:

```python
from collections import Counter

grades=[83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Group 100s with 90s bucket
histogram=Counter(min(grade // 10 * 10, 90) for grade in grades)
print(histogram)

# x + 5 used to center the bar.
plt.bar([x + 5 for x in histogram.keys()], histogram.values(), width=10, edgecolor=(0,0,0))
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of students')
plt.title('Distribution of Exam 1 Grades')
plt.show()
```

### More on Line Charts

These are especially good for showing *trends*.  The following graph will illustrate the Bias-Variance Tradeoff.

```python
variance=[1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared=[256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error=[x + y for x, y in zip(variance, bias_squared)]
xs=[i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9)
plt.xlabel('model complexity')
plt.xticks([])
plt.title('The Bias-Variance Tradeoff')
plt.show()
```

### Simple Scatterplot

Scatterplots are useful for showing relationships or potential correlations between two paired sets of data (or more, but that requires higher dimensions).  Below will illistrate the relationship between height and weight on a sample of male students.

```python
height = [172.5, 165, 170, 172.5, 172.5]
weight = [140, 135, 135, 120, 125]
labels = ['a', 'b', 'c', 'd', 'e']

plt.scatter(height, weight)
for label, height_cms, weight_lbs in zip(labels, height, weight):
    plt.annotate(label, xy=(height_cms, weight_lbs), xytext=(5,-5), textcoords='offset points')

plt.title('Height vs. Weight')
plt.xlabel('weights')
plt.ylabel('heights')
plt.axis('equal')
plt.show()
```
